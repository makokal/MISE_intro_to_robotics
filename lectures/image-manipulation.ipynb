# Capturing images from Camera

We learn get images from either a system camera or external one via `opencv`

```python
camera = cv2.VideoCapture(0)  # Number for the camera
ret, image = camera.read()  # Get the latest image 
#Photoshop by hand!
image[550:650, 500:550, 2] = 255  # R
image[550:650, 500:550, 1] = 155  # G
image[550:650, 500:550, 0] = 0  # B
plt.imshow(image[:, :, ::-1])
```

```python
camera.release()
```

## Zooming

```python
lama = img[400:620,450:700,:]
head = lama[10:50, 5:50,:]
plt.imshow(head[:, :,::-1])
```

## Scaling

```python
scaled_head = cv2.resize(
    head,None,fx=.2, fy=.2, 
    interpolation = cv2.INTER_CUBIC)

scaled_down = cv2.resize(
    scaled_head,None,fx=5, fy=5, 
    interpolation = cv2.INTER_CUBIC)

#OR

height, width = head.shape[:2]
print(height, width)
```

```python
plt.imshow(scaled_head[:, :,::-1])
```

```python
plt.imshow(scaled_down[:, :,::-1])
```

## Translation

```python
rows,cols,channels = lama.shape
M = np.float32([[1,0,100],[0,1,-50]])
dst = cv2.warpAffine(lama,M,(cols,rows))
plt.imshow(dst[:, :,::-1])
```

## Rotation

```python
rows,cols,channels = lama.shape

M = cv2.getRotationMatrix2D((0, 0),330,.5)
dst = cv2.warpAffine(lama,M,(cols,rows))
plt.imshow(dst[:, :,::-1])
```
